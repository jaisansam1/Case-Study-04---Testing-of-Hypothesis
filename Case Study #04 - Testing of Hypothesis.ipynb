```python
from scipy.stats import chi2
chi_square = sum([(0-e)**2./e for o,e in zip(observed_values,expected_values)])
chi_square_statistic = chi_square[0]+chi_square[1]+chi_square[2]
```

```python
print("chi-square statistic: ",chi_square_statistic)
```

```python
critical_value = chi2.ppf(q=1-alpha,df=degree_of_freedom)
print('critical_value: ',critical_value)
```

# The above values are enough to check the dependancy, however to ensure our hypothesis is correct - Calculating p value as an extra measure of check

```python
p_value = 1-chi2.cdf(x=chi_square_statistic,df=degree_of_freedom)
print('p_value: ', p_value)
print('significance level: ', alpha)
print('degree of freedom: ', degree_of_freedom)
```

# ALTERNATE HYPOTHESIS: Definite dependancy between Region and Manager
